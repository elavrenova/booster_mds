```python
import numpy as np
import pandas as pd
```

```python
#Reading data
test_data = pd.read_csv('coursera_sessions_test.txt', delimiter=';', names=['looked','bought'])
train_data = pd.read_csv('coursera_sessions_train.txt', delimiter=';', names=['looked','bought'])
test_data.head()
```

### 1. Подготовительные работы

```python
def get_freq_list(data):
    l = list()
    for string in data:
        id_list = string.strip().split(',')
        l += id_list
    l = list(map(int, l))
    frequency_qty = np.bincount(l)
    print(frequency_qty)
    non_zero_ids = np.nonzero(frequency_qty)[0]
    print(non_zero_ids)
    return list(zip(non_zero_ids, frequency_qty[non_zero_ids]))
```

```python
# looks
train_looks = get_freq_list(list(train_data['looked']))
train_looks_data = pd.DataFrame(train_looks, columns=['id','looks']).set_index('id')

# buys
train_data_b = train_data['bought'].dropna()
train_buys = get_freq_list(list(train_data_b))
train_buys_data = pd.DataFrame(train_buys, columns=['id','buys']).set_index('id')
```

```python
train_looks_buys_data = train_looks_data.join(train_buys_data)
train_looks_buys_data= train_looks_buys_data.fillna(0)
train_looks_buys_data

train_looks_buys_data['looked_rank'] = train_looks_buys_data.looks.rank(ascending=False)
train_looks_buys_data['bought_rank'] = train_looks_buys_data.buys.rank(ascending=False)
train_looks_buys_data
```

### 2. Работа с тестовой выборкой

```python
ids_list = list()
for string in test_data.looked:
    id_list = string.strip().split(',')
    ids_list += id_list
ids_list = list(map(int, ids_list))

test_ids_set = set(ids_list)
dif = test_ids_set.difference(list(train_looks_buys_data.index))
dif_list = list(dif)

dif_data = pd.DataFrame(index = dif_list)
dif_data

train_looks_buys_data = train_looks_buys_data.append(dif_data)
train_looks_buys_data = train_looks_buys_data.fillna(0)
train_looks_buys_data
```

### 3. Задания

```python
def get_metrics(looked, bought, data_type, k):
    bought_qty = 0
    
    looked_list = looked.strip().split(',')
    bought_list = bought.strip().split(',')
    
    looked_data = list(map(int, looked_list))
    bought_data = list(map(int, bought_list))

    l = len(looked_data)
    recom_numb = min(len(looked_data), int(k))
    looked_set = set(looked_data)
    result_list = train_looks_buys_data.loc[looked_set, data_type].sort_values(kind='mergesort')
    recommendations = list(result_list[0:recom_numb].index)
    
    for i in bought_data:
        if i in recommendations:
            bought_qty = bought_qty + 1
            
    precision = bought_qty / k
    recall = bought_qty / len(bought_data)
    return precision, recall

def get_metrics_for_k_array(data, k_array):
    data_rows = data.dropna()
    for row in data_rows.iterrows():
        row_id = row[0]
        row_looked = row[1]['looked']
        row_bougth = row[1]['bought']
        for k in k_array:
            for data_type in ['looked_rank','bought_rank']:
                precision, recall = get_metrics(row_looked, row_bougth, data_type, k)
                data.loc[row_id, 'precision k=' + str(k) + ' '+ data_type] = precision
                data.loc[row_id, 'recall k='+ str(k) + ' ' + data_type] = recall
```

```python
k_array = [1,5]
get_metrics_for_k_array(train_data, k_array)
train_data
```

### 4. Посчитаем средние значения

```python
def get_avg_metrics(data, k_array):
    for k in k_array:
        for metric in ['recall', 'precision']:
            result_string = ''
            for data_type in ['looked_rank','bought_rank']:
                result_string += data_type + ' ' + metric + str(k) + '\t'
                result = data[metric + ' k=' + str(k) + ' ' + data_type].mean(skipna=True)
                rounded_result = round(result, 2)
                result_string += str(rounded_result) + '     '
            print(result_string)
    return
```

```python
get_avg_metrics(train_data, k_array)
```
