```python
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
```

```python
data = pd.read_csv('checkins.dat')
```

```python
data.head()
```

```python
checkins_data = pd.read_csv('checkins.dat', sep='|', skipinitialspace=True, skiprows=[1])
checkins_data.info()
```

```python
print(checkins_data.shape)
checkins_data.head()
```

```python
checkins_data = checkins_data.dropna()
print(checkins_data.shape)
```

```python
checkins_data_cut = checkins_data.iloc[:100000, :]
checkins_data_cut.info()
```

```python
checkins_data_cut.columns = checkins_data_cut.columns.str.strip()
checkins_data = checkins_data_cut[['latitude','longitude']]
checkins_data.shape
```

```python
MS = MeanShift(bandwidth = 0.1)
MS.fit(checkins_data)
```

```python
labels = MS.labels_
print(labels)
```

```python
u_labels, qty_in_labels = np.unique(labels, return_counts=True)
for l, c in zip(u_labels, qty_in_labels):
    print("cluster " + str(l) + " : " + str(c))
```

```python
labels_qty_list = []
for l, c in zip(u_labels, qty_in_labels):
    if c > 15:
        labels_qty_list.append(c)
        
print(labels_qty_list)
print("***************")
print(len(labels_qty_list))
```

```python
centers = MS.cluster_centers_
centers_array = np.empty((len(labels_qty_list), 2))
for i, j in enumerate(labels_qty_list):
    centers_array[i] = centers[i]
    
print(centers_array)
```

```python
cruise_offices = pd.read_csv('cruise_offices.txt')
cruise_offices
```

```python
offices_coordinates = cruise_offices.loc[:, ['latitude', ' longitude']]
offices_coordinates
```

```python
distances = []
for i, c in enumerate(centers_array):
    min_dist = np.sqrt(np.sum((c - offices_coordinates.loc[0])**2))
    for j in range(1, len(offices_coordinates), 1):
        dist = np.sqrt(np.sum((c - offices_coordinates.loc[j])**2))
        if min_dist > dist:
            min_dist = dist
    distances.append(min_dist)
```

```python
distances_sorted = np.argsort(distances)
centers_sorted = centers[distances_sorted]

with open('answer_1.txt', 'w') as file:
    file.write(str(centers_sorted[0, 0]) + ' ' + str(centers_sorted[0, 1]))
```
