```python
# импортируем библиотеки
import numpy as np
from matplotlib import pyplot
import math
from scipy.stats import gamma
import scipy.stats as sts
```

Было выбрано гамма-распределение, подробнее познакомиться можно по ссылке https://en.wikipedia.org/wiki/Gamma_distribution
Но при задании параметра k = 1 данное распределение вырождается в экспоненциальное, все расчеты были проведены для него.

```python
# Задаем параметр a для определения формы графика
a = 1
fig, ax = pyplot.subplots(1, 1)

# Строим гистрограмму на основе сгенерированных данных
r = gamma.rvs(a, size=1000)
ax.hist(r, bins = 20, density=True, histtype='stepfilled', alpha=0.2, label='gamma histogram')

# Выводим график теоретической функции распределения
x = np.linspace(gamma.ppf(0.01, a), gamma.ppf(0.99, a), 1000)
ax.plot(x, gamma.pdf(x, a), 'r-', lw=2, alpha=0.7, label = "gamma pdf")

# Подписываем оси и отображаем график
pyplot.xlabel('$x$')
pyplot.ylabel('$f(x)$')
ax.legend(loc='best', frameon=False)
pyplot.show()
```

## Среднее и дисперсия гамма-распределения

Для гамма-распределения мат.ожидание и дисперсия равны: 
$${\displaystyle \mathbb {E} [X]=k{\theta }}$$ , $${\displaystyle \mathbb {D} [X]=k{\theta ^{2}}}$$

При использовании библиотеки параметр k называется a, а \theta = 1.

Для экспоненциального распределения мат.ожидание и дисперсия равны:

$${\mathbb  {E}}[X]={\frac  {1}{\lambda }}$$ ,
$$\operatorname {D}[X]={\frac  {1}{\lambda ^{2}}} $$

## Применение ЦПТ

```python
# Для заданных значений объема выборки
for n in [5, 10, 50]:
    
    # Генерируем список выборочных средних
    samples_list = [gamma.rvs(a, size=n).mean() for i in range(1000)]
    
    # Строим гистограмму для новой выборки
    pyplot.hist(samples_list, bins = 20, density=True, histtype='stepfilled', alpha=0.2, label='histogram')
    
    # Строим график плотности норм.распределения
    x = np.linspace(0, 4, 100)
    mu = a # mu = a
    std = np.sqrt(1/n) # стандартное отклонение
    normal_distr = sts.norm.pdf(x , mu, std)
    pyplot.plot(x, normal_distr, 'r-', lw=2, label='norm distr')
    
    # Подписываем оси и отображаем график
    pyplot.xlabel('$x$')
    pyplot.ylabel('$f(x)$')
    pyplot.legend(loc='best', frameon=False)
    pyplot.show()
```

## Вывод

С ростом n качество аппроксимации увеличивается, то есть выборочное распределение все больше приближается к восстанавливаемому нормальному распределению.
